```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
chrome_options = Options()
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument("test-type")
chrome_options.add_argument("start-maximized")
chrome_options.add_argument("--window-size=1920,1080")
chrome_options.add_argument("--enable-precise-memory-info")
chrome_options.add_argument("test-type=browser")
driver = webdriver.Chrome(chrome_options=chrome_options,executable_path='C:/Users/BUIQUOCHUY/Desktop/CODE-STUDY/BA/Stock-Prediction/Bot_Craw_data/chromedriver_win32/chromedriver')
link_base='https://vn.investing.com/indices/us-30-news/'
```

```python

max_page=1770//10
list_link=[]

for i in range(1,max_page):
    link_page=link_base+str(i)
    list_link.append(link_page)
```

```python
# get_attribute('innerHTML')
# driver.find_element_by_tag_name('a')
for i in range(len(list_link)):
    url=list_link[i]
    driver.get(url)
    body=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]')
    list_new_apage=body.find_elements_by_class_name('articleItem')
    for index in range(len(list_new_apage)):
        try:
            try:
                a=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]').find_elements_by_class_name('articleItem')[index].find_element_by_tag_name('img')
            except:
                a=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]').find_elements_by_class_name('articleItem')[index].find_element_by_tag_name('a')
            a.click()
            print(driver.find_element_by_class_name('articleHeader').text)
            driver.execute_script("window.history.go(-1)")
        except:
            print("error")  
            continue
    break
```
