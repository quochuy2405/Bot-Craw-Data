```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import re
import time
import pandas as pd
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
driver = webdriver.Chrome(chrome_options=chrome_options,executable_path='C:/Users/BUIQUOCHUY/Desktop/CODE-STUDY/BA/Stock-Prediction/Bot_Craw_data/chromedriver_win32/chromedriver')
link_base='https://vn.investing.com/indices/us-30-news/'
```

```python


list_link=[]

for i in range(20,100):
    link_page=link_base+str(i)
    list_link.append(link_page)
```

```python
data=[]
for i in range(len(list_link)):
   try:
      url=list_link[i]
      driver.get(url)
      
      for count in range(0,10,1):
         try:
            try:
               link=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]').find_elements_by_class_name('articleItem')[count].find_element_by_class_name('title')
            except:
               link=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]').find_elements_by_class_name('articleItem')[count].find_element_by_class_name('img')
            title=link.text
            link.click()
            dates=driver.find_element_by_xpath('/html/body/div[4]/section/div[1]/span').text
            date_to=re.search(r'\d{2}/\d{2}/\d{4}', str(dates))
            date=date_to.group(0)
            content=driver.find_element_by_xpath('/html/body/div[4]/section/div[3]').text

            data.append([date,title, content])
            time.sleep(1)
            driver.execute_script("window.history.go(-1)")
         except:
            None
   except:
      print(i)
      break
df = pd.DataFrame(data,columns=['Date','Title','Content'])
print(df)
df.to_excel("output_part3.xlsx")  
```
