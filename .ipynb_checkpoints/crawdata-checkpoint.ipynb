```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import re
import pandas as pd
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--window-size=1920x1080")
driver = webdriver.Chrome(chrome_options=chrome_options,executable_path='C:/Users/BUIQUOCHUY/Desktop/CODE-STUDY/BA/Stock-Prediction/Bot_Craw_data/chromedriver_win32/chromedriver')
link_base='https://vn.investing.com/indices/us-30-news/'
```

```python

max_page=1770//10
list_link=[]

for i in range(1,max_page):
    link_page=link_base+str(i)
    list_link.append(link_page)
```

```python
# get_attribute('innerHTML')
# driver.find_element_by_tag_name('a')
regex="/^\d{2}\/\d{2}\/\d{4}$/"
data = {
        "Dates": [],
        "Title": [],
        "Contents":[]
}
for i in range(len(list_link)):
    url=list_link[i]
    driver.get(url)
    body=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]')
    list_new_apage=body.find_elements_by_class_name('articleItem')
    for index in range(len(list_new_apage)):
        try:
          
            try:
                link_click=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]').find_elements_by_class_name('articleItem')[index].find_element_by_tag_name('img')
            except:
                link_click=driver.find_element_by_xpath('/html/body/div[4]/section/div[8]').find_elements_by_class_name('articleItem')[index].find_element_by_tag_name('a')
            link_click.click()
            # print(driver.find_element_by_class_name('articleHeader').text)
            content_date=driver.find_element_by_class_name('contentSectionDetails')
            dates=content_date.find_element_by_tag_name('span').text
            date_to=re.search(r'\d{2}/\d{2}/\d{4}', str(dates))
            date=date_to.group(0)
            title=driver.find_element_by_class_name('articleHeader').text
            contents=driver.find_element_by_class_name('articlePage').text
            data['Dates'].append(date)
            data['Title'].append(title)
            data['Contents'].append(contents)
            driver.execute_script("window.history.go(-1)")
        except:
            print("error")  
            continue
        df = pd.DataFrame(data)
        print(df) 
        break
    break
```
